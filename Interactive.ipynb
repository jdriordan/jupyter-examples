```python
%matplotlib inline                 
from ipywidgets import interact    
import numpy as np                 
import matplotlib.pyplot as plt    
```

```python
def plot_test(a=1,b=1,c=1):
    plt.figure(figsize=(8,8))
    
    xs=np.linspace(-np.pi,np.pi,100)
    ys=np.sin(a*xs+b)
    
    plt.plot(xs,ys,
             label="y=sin({a}*x+{b})".format(a=a,b=b)
    )
    plt.plot(xs,(1+ys)**c-c,
            label="(1+y)^{c}-{c}".format(c=c)
    )
    
    plt.legend(loc="upper right")
    plt.show()
    
plot_test()
```

```python
interact(
    plot_test,
    a=(0.1,10.),
    b=(-1,1.),
    c=(1,3.)
)
```
